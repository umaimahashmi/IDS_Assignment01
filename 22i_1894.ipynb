```python
from PIL import Image
import numpy as np
from IPython.display import display
```

Opening the image of chess board and converting our image into an array.

```python
# Open the image
image = Image.open('/content/test.jpg')
display(image)

# Convert the image to a NumPy array
image_array = np.array(image)
```

Defining the ranges for different colors such as black, white and the colors that i want to replace black and white with.

```python
# Define color ranges for black and white in RGB format
black_lower = [0, 0, 0]
black_upper = [70, 70, 70]

white_lower = [200, 200, 200]
white_upper = [255, 255, 255]

# Define the color you want to replace with
replacement_color1 = [0,0,128]  # navy color in RGB
replacement_color2 = [255, 127, 80]  # red color in RGB
```

Converting the black color in the original image to Navy.

```python
# Iterate through each pixel and check if it matches any of the defined color ranges
for y in range(image_array.shape[0]):
    for x in range(image_array.shape[1]):
        pixel = image_array[y, x]

        # Check if pixel matches black color range
        if all(black_lower <= pixel) and all(pixel <= black_upper):
            image_array[y, x] = replacement_color1
```

```python
# Convert the modified NumPy array back to an image
modified_img = Image.fromarray(image_array)

# Display the modified image
display(modified_img)
```

Converting the white color in the original image to Teal.

```python

# Iterate through each pixel and check if it matches any of the defined color ranges
for y in range(image_array.shape[0]):
    for x in range(image_array.shape[1]):
        pixel = image_array[y, x]
        # Check if pixel matches yellow color range
        if all(white_lower <= pixel) and all(pixel <= white_upper):
            image_array[y, x] = replacement_color2
```

Converting the modified array back to image and then displaying it.

```python

# Convert the modified NumPy array back to an image
modified_img = Image.fromarray(image_array)

# Display the modified image
display(modified_img)
```

```python
import pandas as pd
```

```python
file_path = '/timetablemonday.xlsx'
MONDAY= pd.read_excel(file_path)

# Keep columns with at least 50% non-NaN values
# Remove columns where the number of non-NaN values is below the threshold
threshold = len(MONDAY) * 0.5
MONDAY=MONDAY.dropna(axis=1, thresh=threshold)

#giving the data frame new names
new_column_names = ["Room", "8:30-9:50", "10:00-11:20", "11:30-12:50", "1:00-2:20", "2:30-3:30"]

# inserting the day column as the first column
if 'Day' not in MONDAY.columns:
    MONDAY.insert(0, 'Day', 'Monday')
print(MONDAY)
```
